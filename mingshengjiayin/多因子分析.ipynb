```python
def pca(data,n_com):
    data = data.fillna(0)
    pca = PCA(n_components=n_com)
    data_pca = pca.fit_transform(data.iloc[:,:-1])
    tmp = pca.explained_variance_ratio_
    sum = pd.DataFrame(tmp).sum()[0]
    array = tmp
    return sum,array

sum_dic = {}
arrays = []
for i in xrange(5,25):
    tmp = pca(data,i)
    sum_dic[i] = tmp[0]
    arrays.append(list(tmp[1]))
        
print sum_dic
for a in arrays:
    print ','.join([str(i) for i in a])
```
